### Escalado

```python
fit_scal = ScaleFit(data=tdf,
                   target_columns=['VATP009_N','VATP016_N','VATP017_N'],
                   scale_method="RANGE",
                   miss_value="KEEP",
                   global_scale=False)
fit_scal.output
```

```python
## Aplicando la transformación
scal = ScaleTransform(data=tdf, 
                     object=fit_scal.output,
                     accumulate=['LF001'])
```

```python
scal.result.head(10)
```

### Matriz de correlaciones

```python
## Función de VAL para el cálculo In-Database de la matriz de correlaciones
Cor_Mat = valib.Matrix(data=scal.result, columns="allnumeric", exclude_columns=["LF001"], type="COR")
## Imprimir los resultados
Cor_Mat.result.sort('rownum').head(21)
```

```python
## Graficando la Matriz de Correlaciones - Más claro significa correlación más alta
sns.set(rc = {'figure.figsize':(16,12)})
sns.heatmap(Cor_Mat.result.to_pandas().sort_values(['rownum']))
```

### Moviendo los Datos A Python

```python
sub_sample = tbl_train.to_pandas(all_rows=True)
```

```python
sub_sample.head(5)
```

```python
def plot_num(feature):
    fig,axes=plt.subplots(1,2)
    sns.boxplot(data=sub_sample,x=feature,ax=axes[0])
    sns.distplot(a=sub_sample[feature],ax=axes[1],color='#ff4125')
    fig.set_size_inches(15,5)
```

```python
plot_num('VATP009_N')
```

```python
plot_num('VATP016_N')
```

```python
plot_num('VATP017_N')
```

## **Entrenamiento de un Modelo Isolation Forest**

```python
clf = IsolationForest(n_estimators=100, max_samples=100, contamination=0.01,random_state=123, n_jobs=-1)
clf.fit(sub_sample)
```

```python
y_pred = clf.predict(sub_sample)
y_pred.shape
```

```python
unique, counts = np.unique(y_pred, return_counts=True)
dict(zip(unique, counts))
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
scatter = plt.scatter(sub_sample.VATP009_N, sub_sample.VATP017_N, c=y_pred, s=20, edgecolor="k")
handles, labels = scatter.legend_elements()
plt.axis("square")
plt.legend(handles=handles, labels=["outliers", "inliers"], title="true class")
plt.title("Gaussian inliers with \nuniformly distributed outliers")
plt.show()
```

### Exportando el Modelo como PMML

```python
from sklearn2pmml.pipeline import PMMLPipeline
from sklearn2pmml import sklearn2pmml
```

```python
## Se crea el Pipeline para Exportar el modelo en formato PMML
pipeline = PMMLPipeline([("outlier_detector", IsolationForest(n_estimators=100, max_samples=100, contamination=0.01,random_state=123, n_jobs=-1))])
pipeline.fit(sub_sample, y_pred)
sklearn2pmml(pipeline, "IsoForest3.pmml", with_repr = True)
```

## **Scoring del Modelo Isolation Forest de Python utilizando BYOM en la BD**

```python
con.execute("CREATE SET TABLE DW_LANDING.pmml_models (model_id VARCHAR(40) \
            CHARACTER SET LATIN NOT CASESPECIFIC, model BLOB(2097088000)) \
            PRIMARY INDEX ( model_id );")
```

```python
model_bytes = open("IsoForest3.pmml", "rb").read()
con.execute("insert into DW_LANDING.pmml_models  (model_id, model) values(?,?)", 'if3_model', model_bytes)
```

```python
pd.read_sql("select * from DW_LANDING.pmml_models", con)
```

```python
pd.read_sql("select top 10 * from DW_LANDING.Test3IF", con)
```

```python
con.execute("CREATE TABLE DW_LANDING.caso3_ScoreResult AS ( \
SELECT * FROM mldb.PMMLPredict( \
    ON (SELECT * FROM DW_LANDING.Test3IF) \
    ON (select * from DW_LANDING.pmml_models where model_id='if3_model') DIMENSION \
    USING \
        Accumulate('LF001','VATP009_N','VATP016_N','VATP017_N') \
) AS dt \
) WITH DATA;")
```

```python
pd.read_sql("select top 10 * from DW_LANDING.caso3_ScoreResult", con)
```

```python
con.execute("ALTER TABLE DW_LANDING.caso3_ScoreResult ADD flag_out SMALLINT")
```

```python
con.execute("UPDATE DW_LANDING.caso3_ScoreResult SET flag_out=1 WHERE prediction<0")
```

```python
con.execute("UPDATE DW_LANDING.caso3_ScoreResult SET flag_out=0 WHERE flag_out IS NULL")
```
