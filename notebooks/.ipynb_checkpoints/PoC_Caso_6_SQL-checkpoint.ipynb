### Creamos la Matriz de KPIs a nivel DAM y Serie

```
REPLACE VIEW V_KPIS_ADUANAS_ALL AS 
SELECT T1.DAM_RECOD, SERIE, T1.IMPORTADOR, TIPO_UNIDADES_FISICA, AGENTE_ADUANA, DEPOSITO, PAIS_ORIGEN, PARTIDA_ARANCELARIA, 
'C'||SUBSTR(PARTIDA_ARANCELARIA,1,2) CAPITULO, T3.FOB, PESO_NETO, CANTIDAD_UNIDAD_FISICA
FROM c6_AC_DAM_2022_final T1 INNER JOIN c6_Atributos_Importador_2022_final T2 ON T1.IMPORTADOR=T2.IMPORTADOR
INNER JOIN c6_AC_Series_2022_FINAL T3 ON T1.DAM_RECOD=T3.DAM_RECOD
WHERE CAST(FECHA_DAM AS DATE) BETWEEN CURRENT_DATE - INTERVAL '1' YEAR AND CURRENT_DATE;
```

## **Detección de Anomalías**
### Outliers Univariados

```
SELECT * FROM TD_OutlierFilterFit (
    ON V_KPIS_ADUANAS_ALL AS InputTable
    OUT TABLE OutputTable (outlier_fit)
    USING
    TargetColumns ('FOB','PESO_NETO','CANTIDAD_UNIDAD_FISICA')
    LowerPercentile (0.01)
    UpperPercentile (0.99)
    OutlierMethod ('Percentile')
    ReplacementValue ('median')
    PercentileMethod ('PercentileCont')
  ) AS dt;
```

### Aplicando la transformación

```
SELECT top 10 * FROM TD_OutlierFilterTransform (
  ON V_KPIS_ADUANAS_ALL AS InputTable PARTITION BY ANY
  ON outlier_fit AS FitTable DIMENSION
) AS dt;
```

### Outliers Multivariados (Algoritmo OneClassSVM)

```
CREATE multiset table oneclasssvm_model AS (
SELECT * FROM TD_OneClassSVM (
  ON V_KPIS_ADUANAS_ALL AS InputTable
  USING
    InputColumns('FOB','PESO_NETO')
    Tolerance(1e-7)
    BatchSize(30)
    LearningRate('constant')
    InitialEta (0.01)
    RegularizationLambda(0.1)
    Alpha(0)
    Momentum (0.0)
    Nesterov ('false')
    MaxIterNum (150)
) AS dt
) WITH DATA;
```

### Prediction

```
CREATE multiset table prediction AS (
    SELECT * from TD_OneClassSVMPredict(
        ON V_KPIS_ADUANAS_ALL AS InputTable PARTITION BY ANY
        ON oneclasssvm_model AS ModelTable DIMENSION
        USING
        IDColumn('DAM_RECOD')
        Accumulate('SERIE')
        OutputProb('t') 
    ) AS dt
) WITH DATA;
```

### Comparación de Textos de Arancel y Descripción Comercial

```
CREATE MULTISET TABLE c6_aranceles_texto AS (
SELECT T1.DAM_RECOD, T1.SERIE, PARTIDA_ARANCELARIA, 
TRIM(REGEXP_REPLACE(REGEXP_REPLACE(REGEXP_REPLACE(DESCRIPCION_COMERCIAL,'[^a-zA-Z]' ,' '),'  ',' '),'  ',' ')) AS DESCRIPCION_COMERCIAL, 
TRIM(REGEXP_REPLACE(REGEXP_REPLACE(REGEXP_REPLACE(REGEXP_REPLACE(REGEXP_REPLACE(REGEXP_REPLACE(DESCRIPCION_ARANCEL,
'á','a'),'é','e'),'ó','o'),'[^a-zA-Z]' ,' '),'  ', ' '),'  ',' ')) AS DESCRIPCION_ARANCEL
FROM c6_AC_Series_2022_FINAL T1 LEFT JOIN c6_Arancel T2 ON T1.PARTIDA_ARANCELARIA=T2.PARTIDA
LEFT JOIN c6_FormatoB_2022_FINAL T3 ON T1.DAM_RECOD=T3.DAM_RECOD AND T1.SERIE=T3.SERIE) WITH DATA PRIMARY INDEX (DAM_RECOD);
```

```
SELECT COUNT(1) N FROM c6_aranceles_texto;
```

```
SELECT TOP 100 * FROM c6_aranceles_texto;
```

```
CREATE TABLE c6_metrica AS (
SELECT * FROM StringSimilarity (
 ON c6_aranceles_texto PARTITION BY ANY
 USING
 ComparisonColumnPairs ('LD(DESCRIPCION_COMERCIAL,DESCRIPCION_ARANCEL) AS similarity')
 Accumulate ('DAM_RECOD','SERIE','PARTIDA_ARANCELARIA','DESCRIPCION_COMERCIAL','DESCRIPCION_ARANCEL')
) AS dt) WITH DATA;
```

```
SELECT TOP 10 * FROM c6_metrica;
```

```
SELECT COUNT(1) N FROM c6_metrica;
```
