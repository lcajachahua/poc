### Realizando Transformaciones sobre el Dataset

```
CREATE TABLE caso1_onehot as (
SELECT * FROM TD_OneHotEncodingFit(
ON caso1_des AS INPUTTABLE
USING
TargetColumn('VAR_0001','VAR_0002','VAR_0003')
OtherColumnName('other')
IsInputDense('true')
CategoryCounts(3,3,3)
Approach('Auto')
) AS dt 
) with data;
```

```
SELECT * FROM caso1_onehot;
```

```
CREATE TABLE caso1_des_one AS (
SELECT * FROM TD_OneHotEncodingTransform (
ON caso1_des AS InputTable
ON caso1_onehot AS FitTable Dimension
USING
IsInputDense('True')
) AS dt) WITH DATA;
```

```
SELECT TOP 10 * FROM caso1_des_one;
```

```
select * from TD_scaleFit(
on caso1_des_one as InputTable
OUT PERMANENT TABLE OutputTable(scaleFitOut)
using
TargetColumns('VAR_0006','VAR_0007','VAR_0008','VAR_0011','VAR_0012','VAR_0013','VAR_0015','VAR_0016','VAR_0018','VAR_0019','VAR_0020')
MissValue('Keep')
ScaleMethod('range')
GlobalScale('f')
) as dt;
```

```
CREATE TABLE caso1_tf AS (
select * from TD_scaleTransform(
on caso1_des_one as InputTable
on scaleFitOut as FitTable DIMENSION
using
Accumulate('LAB_0002','VAR_0001_1','VAR_0001_2','VAR_0002_1','VAR_0002_2','VAR_0003_1','VAR_0003_2','VAR_0004')
)as dt) WITH DATA;
```

```
SELECT TOP 10 * FROM caso1_tf;
```

```
SELECT * from TD_GLM (
        ON caso1_tf
        USING
        InputColumns('VAR_0001_1','VAR_0001_2','VAR_0002_1','VAR_0002_2','VAR_0003_1','VAR_0003_2','VAR_0006','VAR_0007','VAR_0008','VAR_0011','VAR_0012','VAR_0013','VAR_0015','VAR_0016','VAR_0018','VAR_0019','VAR_0020')
        ResponseColumn('VAR_0004')
        Family('Binomial')
        BatchSize(10)
        MaxIterNum(300)
        RegularizationLambda(0.02)
        Alpha(0.15)
        IterNumNoChange(50)
        Intercept('true')
        LearningRate('invtime')
        InitialEta(0.05)
        Momentum(0)
        Nesterov('false')
        LocalSGDIterations(0)
    ) as dt ORDER BY attribute;
```

### División en Muestras Train y Test

```
-- Muestra de Entrenamiento
CREATE MULTISET TABLE tch_train AS (SELECT * FROM caso1_tf SAMPLE 0.7) WITH DATA
PRIMARY INDEX (LAB_0002);
```

```
-- Muestra de Test
CREATE MULTISET TABLE tch_test AS (SELECT * FROM caso1_tf WHERE LAB_0002 NOT IN (SELECT LAB_0002 FROM tch_train)) WITH DATA
PRIMARY INDEX (LAB_0002);
```

## **Entrenamiento del Modelo (In-Database)**

### Generación del Modelo de Estimación de Precios utilizando Regresión Lineal Múltiple con Selección de Variables Stepwise

```
call val.td_analyze('logistic',
'database=DW_LANDING;
tablename=tch_train;
columns= all;
columnstoexclude=LAB_0002,Y,YL;
dependent=VAR_0004;
statstable=true;
enter=0.001;
remove=0.001;
stepwise=true;
outputdatabase=DW_LANDING;
outputtablename=logisticmodel;');
```

```
-- Parámetros del Modelo
SELECT * FROM logisticmodel ORDER BY 1;
```

```
-- Estadísticas de performance
SELECT * FROM logisticmodel_rpt;
```

## **Evaluación del Modelo**
### Calculando las métricas de diagnóstico del Modelo

```
call val.td_analyze('logisticscore',
'database=DW_LANDING;
tablename=tch_test;
modeldatabase=DW_LANDING;
modeltablename=logisticmodel;
outputdatabase=DW_LANDING;
outputtablename=logisticmodelval;
estimate=estimado;
probability=Probability;
retain=VAR_0004;
samplescoresize=10;
scoringmethod=scoreandevaluate;
lifttable=true;');
```

```
-- Estadísticas de performance
SELECT * FROM logisticmodelval_txt;
```

```
SELECT * FROM TD_ClassificationEvaluator (
   ON (SELECT VAR_0004, CAST(estimado AS INT) estimado FROM logisticmodelval) AS InputTable
   OUT TABLE OutputTable (additional_metrics)
   USING
   ObservationColumn ('VAR_0004')
   PredictionColumn ('estimado')
   Labels (0,1)
) AS dt ORDER BY SeqNum;
```

```
SELECT * FROM additional_metrics;
```
